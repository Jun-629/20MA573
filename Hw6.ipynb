```
import numpy as np
import matplotlib.pyplot as plt  

def exactbm(T,N):
  w_list = [0]                          # set a list with W_0 = 0
  h = T/N                               # set the step
  wi = 0
  for i in range(N):
    Z = np.random.normal(0,1)
    wi += np.sqrt(h) * Z
    w_list.append(wi)
  return w_list

def pltbm(T,N):
  x0 = [x*T/N for x in range(N+1)]
  y0 = exactbm(T,N)
  return plt.plot(x0,y0)

for t in range(100,105):
  pltbm(t,int(t/0.01));
  #plt.plot(t,exactbm(t,int(t/0.01))[int(t/0.01)]/np.sqrt(2*t*np.log(np.log(t))))
  
a = 100
plt.plot(a,np.sqrt(2*a*np.log(np.log(a))),label = 'K_l');
plt.legend();
```
